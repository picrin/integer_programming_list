```python
def action_sa(stacks):
    stack_a, stack_b = stacks
    if len(stack_a) < 2:
        print("error, can't swap stack a, because stack is too short", stack_a)
    new_stack_a = stack_a[:-2] + [stack_a[-1]] + [stack_a[-2]]
    return [new_stack_a, stack_b]

def action_sb(stacks):
    return action_sa(stacks[::-1])[::-1]

def action_ss(stacks):
    return action_sa(action_sb(stacks))

def action_ra(stacks):
    stack_a, stack_b = stacks
    if len(stack_a) < 1:
        print("error, we shouldn't be rotating an empty stack", stack_a)
    new_stack_a = [stack_a[-1]] + stack_a[:-1]
    return [new_stack_a, stack_b]

def action_rb(stacks):
    return action_ra(stacks[::-1])[::-1]

def action_rr(stacks):
    return action_ra(action_rb(stacks))

def action_rra(stacks):
    stack_a, stack_b = stacks
    if len(stack_a) < 1:
        print("error, we shouldn't be reverse rotating an empty stack", stack_a)
    new_stack_a = stack_a[1:] + [stack_a[0]]
    return [new_stack_a, stack_b]

def action_rrb(stacks):
    return action_rra(stacks[::-1])[::-1]

def action_rrr(stacks):
    return action_rra(action_rrb(stacks))

def action_pa(stacks):
    stack_a, stack_b = stacks
    if len(stack_b) < 1:
        print("error, we shouldn't be pushing from an empty stack", stack_b)
    new_stack_b = stack_b[:-1]
    new_stack_a = stack_a + [stack_b[-1]]
    return [new_stack_a, new_stack_b]

def action_pb(stacks):
    return action_pa(stacks[::-1])[::-1]
```

# Is not zero

```python
def is_not_zero_constrain(is_not_zero, state_sum, problem_instance):
    c_1 = pulp.LpConstraint(pulp.LpAffineExpression([(is_not_zero, k), (state_sum,-1)]), pulp.LpConstraintGE, rhs=0)
    c_2 = pulp.LpConstraint(pulp.LpAffineExpression([(is_not_zero, 1), (state_sum,-1)]), pulp.LpConstraintLE, rhs=0)
    problem_instance += c_1
    problem_instance += c_2
```

```python
results = []
k = 4
for is_not_zero_value in [0, 1]:
    for state_sum_value in range(k):
        problem = pulp.LpProblem("check_is_not_zero")
        state_sum, is_not_zero = pulp.LpVariable("state_sum", cat="Integer", lowBound=0, upBound=(k-1)), pulp.LpVariable("is_not_zero", cat="Binary")
        is_not_zero_constrain(is_not_zero, state_sum, problem)
        state_sum.setInitialValue(state_sum_value)
        is_not_zero.setInitialValue(is_not_zero_value)
        state_sum.fixValue()
        is_not_zero.fixValue()
        result = problem.solve()
        results.append([state_sum.value(), is_not_zero.value(), result])
```

# is_lt_two

```python
def is_lt_two_constrain(is_lt_two, state_sum, problem_instance):
    c_1 = pulp.LpConstraint(pulp.LpAffineExpression([(is_lt_two, -k), (state_sum,-1)]), pulp.LpConstraintGE, rhs=(-1-k))
    c_2 = pulp.LpConstraint(pulp.LpAffineExpression([(is_lt_two, -k), (state_sum,-1)]), pulp.LpConstraintLE, rhs=-2)
    problem_instance += c_1
    problem_instance += c_2
```

```python
results = []
k = 3
for is_lt_two_value in [0, 1]:
    for state_sum_value in range(k):
        problem = pulp.LpProblem("check_is_lt_two")
        state_sum, is_lt_two = pulp.LpVariable("state_sum", cat="Integer", lowBound=0, upBound=(k-1)), pulp.LpVariable("is_not_one", cat="Binary")
        is_lt_two_constrain(is_lt_two, state_sum, problem)
        state_sum.setInitialValue(state_sum_value)
        is_lt_two.setInitialValue(is_lt_two_value)
        state_sum.fixValue()
        is_lt_two.fixValue()
        result = problem.solve()
        results.append([state_sum.value(), is_lt_two.value(), result])
```

# Is one

We need to put together constraints: is_lt_two, is_not_zero and the conjuction constraint. We're hoping that the solver can easily deal with the intermediate variables.

```python
def conjuction_constrain(var1, var2, var1_and_var2, problem_instance):
    c_5 = pulp.LpConstraint(pulp.LpAffineExpression([(var1, 1), (var2,1), (var1_and_var2,-1)]), pulp.LpConstraintLE, rhs=1)
    c_6 = pulp.LpConstraint(pulp.LpAffineExpression([(var1, 1), (var2,1), (var1_and_var2,-2)]), pulp.LpConstraintGE, rhs=0)
    problem_instance += c_5
    problem_instance += c_6
```

```python
results = []
k = 5
for is_one_value in [0, 1]:
    for state_sum_value in range(k):
        problem = pulp.LpProblem("check_is_lt_two")
        
        state_sum, is_lt_two = pulp.LpVariable("state_sum", cat="Integer", lowBound=0, upBound=(k-1)), pulp.LpVariable("is_not_one", cat="Binary")
        is_lt_two_constrain(is_lt_two, state_sum, problem)
        
        is_not_zero = pulp.LpVariable("is_not_zero", cat="Binary")
        is_not_zero_constrain(is_not_zero, state_sum, problem)

        is_one = pulp.LpVariable("is_one", cat="Binary")
        conjuction_constrain(is_lt_two, is_not_zero, is_one, problem)
        
        state_sum.setInitialValue(state_sum_value)
        is_one.setInitialValue(is_one_value)
        state_sum.fixValue()
        is_one.fixValue()
        result = problem.solve()
        results.append([state_sum.value(), is_one.value(), result])
```

```python
results
```

# not_a_and_b

```python
def not_a_and_b_constrain(var1, var2, not_var1_and_var2, problem_instance):
    c_5 = pulp.LpConstraint(pulp.LpAffineExpression([(var1, -1), (var2,1), (not_var1_and_var2, -1)]), pulp.LpConstraintLE, rhs=0)
    c_6 = pulp.LpConstraint(pulp.LpAffineExpression([(var1, -1), (var2,1), (not_var1_and_var2, -2)]), pulp.LpConstraintGE, rhs=-1)
    problem_instance += c_5
    problem_instance += c_6
```

```python
results = []
for a_value in range(2):
    for b_value in range(2):
        for c_value in range(2):
            problem = pulp.LpProblem("check_concunction")
            a = pulp.LpVariable("a", cat="Binary")
            b = pulp.LpVariable("b", cat="Binary")
            c = pulp.LpVariable("c", cat="Binary")
            not_a_and_b_constrain(a, b, c, problem)
            a.setInitialValue(a_value)
            b.setInitialValue(b_value)
            c.setInitialValue(c_value)
            a.fixValue()
            b.fixValue()
            c.fixValue()
            result = problem.solve()
            results.append([a.value(), b.value(), c.value(), result])
```

```python
problem_input = [5, 2, 3, 4, 1]
steps = 3
steps_range = [i for i in range(steps)]

problem = pulp.LpProblem("minimise_steps_to_sort_list")

is_swapped = pulp.LpVariable.dicts("is_swapped", steps_range, cat="Binary")
is_rotated = pulp.LpVariable.dicts("is_rotated", steps_range, cat="Binary")
is_reverse_rotated = pulp.LpVariable.dicts("is_reverse_roated", steps_range, cat="Binary")

length_range = [i for i, _ in enumerate(problem_input)]

is_not_zero = pulp.LpVariable.dicts("is_not_zero", [(i, j) for i in steps_range for j in length_range], cat="Binary")
is_lt_two = pulp.LpVariable.dicts("is_lt_two", [(i, j) for i in steps_range for j in length_range], cat="Binary")
is_not_last = pulp.LpVariable.dicts("is_not_last", [(i, j) for i in steps_range for j in length_range], cat="Binary")

is_one = pulp.LpVariable.dicts("is_one", [(i, j) for i in steps_range for j in length_range], cat="Binary")
is_swapped_zero = pulp.LpVariable.dicts("is_swapped_zero", [(i, j) for i in steps_range for j in length_range], cat="Binary")
is_swapped_one = pulp.LpVariable.dicts("is_swapped_one", [(i, j) for i in steps_range for j in length_range], cat="Binary")
is_positive_overflow = pulp.LpVariable.dicts("is_positive_overflow", [(i, j) for i in steps_range for j in length_range], cat="Binary")
is_negative_overflow = pulp.LpVariable.dicts("is_negative_overflow", [(i, j) for i in steps_range for j in length_range], cat="Binary")

initial_state_sum = length_range

final_state_sum = [elem[1] for elem in sorted([(elem, i) for i, elem in enumerate(problem_input)])]

state_sum = pulp.LpVariable.dicts("is_one", [(i, j) for i in steps_range + [len(steps_range)] for j in length_range], cat="Binary")

for i in range(steps):
    # restrict to only one move per step
    one_move_per_step = pulp.LpConstraint(pulp.LpAffineExpression([(is_swapped[i], 1), (is_rotated[i], 1), (is_reverse_rotated[i], 1)]), pulp.LpConstraintEQ, rhs=1)
    problem += one_move_per_step

for j in range(len(problem_input)):
    first_column_state_sum_constraint = pulp.LpConstraint(pulp.LpAffineExpression([(state_sum[(0, j)], 1)]), pulp.LpConstraintEQ, rhs=initial_state_sum[j])
    problem += first_column_state_sum_constraint
    last_column_state_sum_constraint = pulp.LpConstraint(pulp.LpAffineExpression([(state_sum[(steps, j)], 1)]), pulp.LpConstraintEQ, rhs=final_state_sum[j])
    problem += last_column_state_sum_constraint

def state_sum_constrain(state_sum, state_sum_next, is_swapped_zero, is_swapped_one, is_rotated, is_positive_overflow, is_reverse_rotated, is_negative_overflow, problem_instance):
    constraint = pulp.LpConstraint(pulp.LpAffineExpression([(state_sum, 1), (is_swapped_zero, 1), (is_swapped_one, -1), (is_rotated, 1), (is_positive_overflow, -k), (is_reverse_rotated, -1), (is_negative_overflow, k), (state_sum_next, -1)]), pulp.LpConstraintEQ, rhs=0)
    problem_instance += constraint

for i in range(steps):
    for j in range(len(problem_input)):
        is_not_zero_constrain(is_not_zero[(i, j)], state_sum[(i, j)], problem)
        is_lt_two_constrain(is_lt_two[(i, j)], state_sum[(i, j)], problem)
        is_not_last_constrain(is_not_last[(i, j)], state_sum[(i, j)], problem)
        conjuction_constrain(is_lt_two[(i, j)], is_not_zero[(i, j)], is_one[(i, j)], problem)
        not_a_and_b_constrain(is_not_zero[(i, j)], is_swapped[i], is_swapped_zero[(i, j)], problem)
        conjuction_constrain(is_one[(i, j)], is_swapped[i], is_swapped_one[(i, j)], problem)
        not_a_and_b_constrain(is_not_last[(i, j)], is_rotated[i], is_positive_overflow[(i, j)], problem)
        not_a_and_b_constrain(is_not_zero[(i, j)], is_reverse_rotated[i], is_negative_overflow[(i, j)], problem)
        state_sum_constrain(state_sum[(i, j)], state_sum[(i + 1, j)], is_swapped_zero[(i, j)], is_swapped_one[(i, j)], is_rotated[i], is_positive_overflow[(i, j)], is_reverse_rotated[i], is_negative_overflow[(i, j)], problem)
```
